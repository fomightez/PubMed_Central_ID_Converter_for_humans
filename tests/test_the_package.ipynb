## Running the tests of `PubMed_Central_ID_Converter_for_humans`

Now with the set-up done we can run the test file by running the next code cell below.  
Originally, this code in the cell below was what I was using (I have made it `raw` now) ...

Note that while that command works, it leaves in the output weird text ike `]9;4;3;\` and `]9;4;1;10\` that you don't see when running `pytest --no-header -s -vv tests/test_Converter.py` in a terminal window launched from the root of this session. (You can run it here if you want to see.)

Luckly, that weird text cruft (OSC (Operating System Command) sequences for modern terminals) can be filtered out using this code below that is really just running the equivalent of `!pytest --no-header -s -vv tests/test_Converter.py` but filtering the result to leave them out and let it display as if run in the terminal. (As a result of needing to filer out the captured text, you don't see the tests as they progress; it only shows the results after finished. You can just go ahead and run `!pytest --no-header -s -vv tests/test_Converter.py` if you prefer that and don't mind the extra, weird cruft.)  
(Make sure `use_freshest_in_session_package_code` set to `True` in the code below if you are testing code you actively changed since this MyBinder session started.)

```python
import subprocess
import re

#***-----------------For VERY ACTIVE DEVELOPMENT STAGES--------------------------***#
# Note `conftest.py` CURRENTLY only runs `uv pip install -e .` if the `PMC_id_convert` CLI
# command doesn't exist already. If the package was installed when the session created, the 
# CLI command will already exist and that won't be re-run.
# However, you may still be actively developing things and want it to use the most current version
# of the package code in the session. In other words, the session may have started and then
# you changed things to test out pending code changes, and want to ensure the actual MOST 
# current version of the package code is used. 
# If so, make sure to set `use_freshest_in_session_package_code` to `True`
use_freshest_in_session_package_code = True # I plan to set this to `False` when this package
# is more mature because by default, I want the tests to test the current version available 
# via `pip install`; however, I want a way to test things altered after session started, too.
if use_freshest_in_session_package_code:
    subprocess.run(['uv', 'pip', 'install', '-e', '.'], 
                  check=True, 
                  capture_output=True)
    # CRITICAL: Invalidate import caches so Python knows about the new package
    import importlib
    importlib.invalidate_caches()
    # Also ensure the package location is in sys.path
    import site
    importlib.reload(site)
#***--------END ADDITION 'For VERY ACTIVE DEVELOPMENT STAGES' section------------***#
# NOW TO RUN THE TESTS & filter weird text cruft that shows up in Jupyter run ...

result = subprocess.run(
    ['pytest', '-s', '-vv', 'tests/test_Converter.py'], 
    capture_output=True, 
    text=True
)

# Remove OSC sequences (]9;...\) and ANSI escape codes
clean = re.sub(r'\]9;[^\]]*\\', '', result.stdout + result.stderr)
print(clean)
```
